```julia
using MCTS
using POMDPs
using POMDPModels
using D3Trees
using Random
```

```julia
# This defines how nodes in the tree view are labeled.
MCTS.node_tag(s::GWPos) = "[$(s[1]),$(s[2])]"
MCTS.node_tag(a::Symbol) = "[$a]"
```

```julia
n_iter = 100000
depth = 10
ec = 10.0

solver = MCTSSolver(n_iterations=n_iter,
    depth=depth,
    exploration_constant=ec,
    enable_tree_vis=true
)
mdp = SimpleGridWorld()

planner = solve(solver, mdp)
state = rand(Random.MersenneTwister(4), initialstate(mdp))

a = action(planner, state)
```

```julia
# Note, this does not appear correctly on github... to see it, either
#     1. Go to https://nbviewer.jupyter.org/github/JuliaPOMDP/MCTS.jl/blob/master/notebooks/Test_Visualization.ipynb
#     OR
#     2. Run it on your local Machine
a, info = action_info(planner, state)
D3Tree(info[:tree], init_expand=2) # click on the node to expand it
```

```julia
n_iter = 100000
depth = 10
ec = 10.0

solver = DPWSolver(n_iterations=n_iter, depth=depth, exploration_constant=ec, alpha_state=1/8, tree_in_info=true)
mdp = InvertedPendulum()

planner = solve(solver, mdp)
state = rand(MersenneTwister(4), initialstate(mdp))

a, info = action_info(planner, state, tree_in_info=true);
```
